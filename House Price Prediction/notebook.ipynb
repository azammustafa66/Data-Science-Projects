```python
path = 'data/Bengaluru_House_Data.csv'

data = pd.read_csv(path)
```

```python
data.drop(columns=['area_type', 'availability', 'society', 'balcony'], inplace=True)
```

```python
data.describe()
```

```python
data.info()
```

```python
data['location'] = data['location'].fillna('Sarjapur Road')
```

```python
data['size'].value_counts()
```

```python
data['size'] = data['size'].fillna('2 BHK')
```

```python
data['bath'] = data['bath'].fillna(data['bath'].median())
```

```python
data.info()
```

```python
data['bhk'] = data['size'].apply(lambda x: int(x.split(' ')[0])).astype(int)
```

```python
data[data.bhk > 10]
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    try:
        return float(x)
    except:
        return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(convert_sqft_to_num)
```

```python
data.head()
```

Price Per Square feet

```python
data['price_per_sqft'] = data['price'] * 100000 / data['total_sqft']
```

```python
data.describe()
```

```python
data['location'] = data['location'].apply(lambda x: x.strip())
location_stats = data['location'].value_counts()
```

```python
location_stats
```

```python
location_stats_less_than_20 = location_stats[location_stats < 20]
```

```python
location_stats_less_than_20
```

```python
data['location'] = data['location'].apply(lambda x: 'other' if x in location_stats_less_than_20 else x)
```

Outlier detection and removal

```python
data.describe()
```

```python
(data['total_sqft'] / data['bhk']).describe()
```

```python
data = data[(data['total_sqft'] / data['bhk']) >= 300]
data.describe()
```

```python
data['price_per_sqft'].describe()
```

```python
def remove_sqft_outliers(df: pd.DataFrame):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby("location"):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        
        reduced_df = subdf[
            (subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))
        ]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

data = remove_sqft_outliers(data)
data.describe()
```

```python
def remove_bhk_outliers(df: pd.DataFrame):
    exclude_indices = np.array([])
    for _, location_df in df.groupby("location"):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby("bhk"):
            bhk_stats[bhk] = {
                "mean": np.mean(bhk_df.price_per_sqft),
                "std": np.std(bhk_df.price_per_sqft),
                "count": bhk_df.shape[0],
            }
            
        for bhk, bhk_df in location_df.groupby("bhk"):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats["count"] > 5:
                exclude_indices = np.append(
                    exclude_indices,
                    bhk_df[bhk_df.price_per_sqft < stats["mean"]].index.values,
                )
    return df.drop(exclude_indices, axis="index")

data = remove_bhk_outliers(data)
```

```python
data = data.drop(columns=['size', 'price_per_sqft'])
file_path = 'data/Bengaluru_House_Data_Cleaned.csv'
data.to_csv(file_path, index=False)
```

```python
X = data.drop(columns=['price'])
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ["location"]), remainder="passthrough"
)
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression(fit_intercept=True)
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
