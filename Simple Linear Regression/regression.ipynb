# Simple Linear Regression: Predicting TV Sales

In this notebook, we will perform a simple linear regression analysis to predict TV sales based on advertising expenditure. We will use a dataset that contains information on TV advertising budgets and corresponding sales figures. Our goal is to understand the relationship between advertising spend and sales, and to build a model that can predict future sales based on advertising budgets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Loading the Dataset

First, we need to load the dataset containing the TV advertising budgets and sales figures. We'll open the CSV file and read it into a pandas DataFrame for easy manipulation and analysis.


```python
path = 'data/tvmarketing.csv'
data = pd.read_csv(path)
```

```python
data.head()
```

### Plot the graph

```python
plt.plot(data['TV'], data['Sales'], 'o', label='TV Sales', color='black')
plt.xlabel('TV Advertising Budget (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.title('TV Advertising vs. Sales')
plt.legend()
plt.show()
```

## Calculating the Line of Best Fit

To quantify the relationship between TV advertising budgets and sales, we'll use linear regression to find the line of best fit. This involves calculating the slope and intercept of the regression line.


```python
m, b = np.polyfit(data['TV'], data['Sales'], 1)
print(f'Slope: {m}, Intercept: {b}')
```

## Plotting the Best Fit Line

Once we have the slope and intercept, we'll plot the line of best fit on the graph. This will help us see how well the line represents the data.


```python
plt.plot(data['TV'], data['Sales'], 'o', label='TV Sales', color='black')
plt.plot(data['TV'], m*data['TV']+b, '-', label='Best Fit Line', color='red')
plt.xlabel('TV Advertising Budget (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.title('TV Advertising vs. Sales')
plt.legend()
plt.show()
```

## Making Predictions

Finally, we can use our regression model to make predictions. We'll take some test values of TV advertising budgets and predict the corresponding sales figures.


```python
test_sales = np.array([50, 120, 150])
predicted_sales = m * test_sales + b

print(f'Predicted Sales for $50k: {predicted_sales[0].round(5)}')
print(f'Predicted Sales for $120k: {predicted_sales[1].round(5)}')
print(f'Predicted Sales for $150k: {predicted_sales[2].round(5)}')
```
